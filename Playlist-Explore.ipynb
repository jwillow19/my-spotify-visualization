```python
# Find all artists and count their occurance
artists = [track['by'] for track in top2019.tracks]; artists

def count_by_artist(artists):
    artist_dict = {}
    
#     Logic - loop through the list of artists, check if artist is in dict
# if false: add to dict
# if true: increment value by 1
    for artist in artists:
        if artist not in artist_dict:
            artist_dict[artist] = 1
            pass
        num_occurance = artist_dict[artist]
        to_update = {artist:num_occurance+1}
        artist_dict.update(to_update)
    
    return artist_dict 
        
            
```

```python
artist_dict = count_by_artist(artists)

# sort by artist with most tracks
sorted_artists = dict(sorted(artist_dict.items(), key=lambda kv: kv[1], reverse=True))

print('Number of unqiue artists: ', len(artist_dict.keys()))
sorted_artists
```
